# COVID-19 PROJECT

```python
# import the important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset
data = pd.read_csv('/content/covid_19_clean_complete.csv')
data.head()
```

```python
# Rename the country & state column
data.rename(columns={'Province/State':'state','Country/Region':'country'},inplace=True)
data.head()
```

```python
# Data info
data.info()
```

```python
data.Date.value_counts()
```

```python
# Shape of the dataset
data.shape
```

```python
data.columns
```

```python
# Let's grouping all the confirmed case by Date
Confirmed = data.groupby(by=['Date'])['Confirmed'].sum().reset_index()
Confirmed
```

Plotting the `confirmed case`

```python
plt.figure(figsize=(20,8))
sns.lineplot(x='Date',y='Confirmed',data=Confirmed)
plt.title('Confirmed Cases in the World')
plt.show()
```

```python
Deaths = data.groupby(by=['Date'])['Deaths'].sum().reset_index()
Deaths
```

Plotting the `Death cases`

```python
plt.figure(figsize=(20,8))
sns.lineplot(x='Date',y='Deaths',data=Deaths)
plt.title('Death Cases in the World')
plt.show()
```

```python
Recovered = data.groupby(by=['Date'])['Recovered'].sum().reset_index()
Recovered
```

Plotting the `Recovered cases`

```python
plt.figure(figsize=(20,8))
sns.lineplot(x='Date',y='Recovered',data=Recovered)
plt.title('Recovered Cases in the World')
plt.show()
```

```python
Active = data.groupby(by=['Date'])['Active'].sum().reset_index()
Active
```

plotting the `Active cases`

```python
plt.figure(figsize=(20,8))
sns.lineplot(x='Date',y='Active',data=Active)
plt.title('Active Cases in the World')
plt.show()
```

```python
plt.subplots(2,2,figsize=(20,8))
plt.subplot(2,2,1)
sns.lineplot(x='Date',y='Confirmed',data=Confirmed)
plt.title('Confirmed Cases in the World')
plt.subplot(2,2,2)
sns.lineplot(x='Date',y='Deaths',data=Deaths)
plt.title('Death Cases in the World')
plt.subplot(2,2,3)
sns.lineplot(x='Date',y='Recovered',data=Recovered)
plt.title('Recovered Cases in the World')
plt.subplot(2,2,4)
sns.lineplot(x='Date',y='Active',data=Active)
plt.title('Active Cases in the World')

plt.tight_layout()
plt.show()
```

## Forecasting for confirmed cases

```python
from prophet import Prophet
```

```python
!pip install --upgrade --no-cache-dir prophet
```

```python
confirm_prophet = confirm.rename(columns={'Date': 'ds', 'Confirmed': 'y'})
model = Prophet()
model.fit(confirm_prophet)
```

```python
future = model.make_future_dataframe(periods=7)
forcast = model.predict(future)
```

```python
forcast.tail(10)
```

```python
forecast_next_7_days = forcast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forecast_next_7_days
```
